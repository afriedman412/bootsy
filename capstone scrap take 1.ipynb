```python
import pandas as pd
import billboard
import wikipedia
import wikipediaapi

w = wikipediaapi.Wikipedia('en')
w_html = wikipediaapi.Wikipedia(language='en', extract_format=wikipediaapi.ExtractFormat.HTML)
```

```python
def chart_hasher(chart, date, df=True):
    c_list = []
    c = billboard.ChartData(chart, date=date)
    for n in range(50):
        entry = {}
        entry['title'] = c[n].title
        entry['artist'] = c[n].artist
        entry['peak'] = c[n].peakPos
        entry['last_week'] = c[n].lastPos
        entry['weeks'] = c[n].weeks
        entry['rank'] = c[n].rank
        entry['new'] = c[n].isNew
        entry['chart_date'] = date
        c_list.append(entry)
    if df:
        c_list = pd.DataFrame(c_list)
    return c_list
```

```python
def wiki_parse(entry):
    w = wikipediaapi.Wikipedia('en')
    w_h = wikipediaapi.Wikipedia(language='en', extract_format=wikipediaapi.ExtractFormat.HTML)
    page = w.page(entry)
    for s in page.sections:
        if 'Personnel' in str(s):
            s_split = str(s).split('\n')
            personnel = {'who':[], 'played':[]}
            for p in s_split[1:]:
                p = p.split('-')
                if len(p) > 1:
                    personnel['who'].append(p[0].strip())
                    personnel['played'].append(p[1].strip())
                else:
                    continue
                    
            return personnel

        else:
            continue
    return
```
